```python
from faker import Faker
import mysql.connector

# Establish connection
def create_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        if connection.is_connected():
            print("Connection to MySQL DB successful")
    except mysql.connector.Error as e:
        print(f"The error '{e}' occurred")
    return connection

connection = create_connection("localhost", "root", "Shaurya@123!", "market")

# Insert records using Faker
fake = Faker()
cursor = connection.cursor()

for _ in range(100):
    name = fake.name()
    email = fake.email()
    phone = fake.phone_number()[:15]  # Restrict phone number length to 15 characters
    
    insert_query = "INSERT INTO student (name, email, phone) VALUES (%s, %s, %s)"
    cursor.execute(insert_query, (name, email, phone))

connection.commit()
print("100 records inserted successfully.")

# Fetch the records from the table
fetch_query = "SELECT * FROM student LIMIT 100"
cursor.execute(fetch_query)
students = cursor.fetchall()

# Display the records
for student in students:
    print(student)

cursor.close()
connection.close()
```
